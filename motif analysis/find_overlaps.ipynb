```python
import pandas as pd
validated = pd.read_csv('experimentally_validated/validated_enhancers.csv')
dev = pd.read_csv('generated_dev.csv')
hk = pd.read_csv('generated_hk.csv')
```

```python
validated = validated[:152]
dev = dev[:162]
hk = hk[:289]
```

```python
def search_for_match(motif_name, consensus, row_num, matches):
    for index, row in validated.iterrows():
        if row['Motif Name'] == motif_name and row['Consensus'] == consensus:
            matches['Motif Name'].append(motif_name) 
            matches['Consensus'].append(consensus) 
            matches['Rank in Generated'].append(row_num + 1)
            matches['Rank in Known'].append(index + 1)
            break
```

```python
d= {'Motif Name': [], 'Consensus': [], 'Rank in Generated': [], 'Rank in Known': []}
for index, row in dev.iterrows():
    search_for_match(row['Motif Name'], row['Consensus'], index, d)
df = pd.DataFrame(d)
df
```

```python
df.to_csv('dev_overlaps.csv')
```

```python
d= {'Motif Name': [], 'Consensus': [], 'Rank in Generated': [], 'Rank in Known': []}
for index, row in hk.iterrows():
    search_for_match(row['Motif Name'], row['Consensus'], index, d)
df = pd.DataFrame(d)
df
```

```python
df.to_csv('hk_overlaps.csv')
```
